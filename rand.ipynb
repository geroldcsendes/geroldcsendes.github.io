```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data={
    'cat': ['aardwark', 'dachshund', '...', 'fox', 'zyzzyva'],
    'prob': [0.01, 0.2, 0, 0.6, 0.02]
})
```

```python
fig, axs = plt.subplots(1, 1, figsize=(6, 4))
sns.barplot(data=df, x='cat', y='prob', color='lightblue', ax = axs)

axs.set_title('Softmax output')
axs.set_xlabel('Vocabulary')
axs.set_ylabel('Probabolity')

plt.show()

plt.savefig
```
